```python
import pandas as pd
```

```python
df = pd.read_csv('All_Data.csv')
```

```python
df.groupby('Pos_1').sum()
```

```python
off = ['Offence', 'Right Winger', 'Left Winger', 'Centre-Forward', 'Attacking Midfield']
deff = ['Right-Back', 'Left-Back', 'Centre-Back', 'Defense', 'Defensive Midfield']

for idx, row in df.iterrows():
  if row['Event_Type'] == "Substitution":
    if row['Pos_1'] in off:
      df.at[idx, 'Note'] = 'Offensive Substitution'
    elif row['Pos_1'] in deff:
      df.at[idx, 'Note'] = 'Defensive Substitution'
    else:
      df.at[idx, 'Note'] = 'Neutral Substitution'
```

```python
df.to_csv('All_Data_Final.csv')
```
